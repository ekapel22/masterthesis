```python
import requests
poms_endpoint = "https://rs.poms.omroep.nl/v1/api/subtitles/"
```

In the first program (Pauw), the first few lines (5) gives a summary of the subjects in the whole episode
In other programs, the first few lines contain a flashback of the last episode.
So a distinction should first be made for type of program. 

Also, in a talkshow you can more easily recommend a random episode that is interesting, but for a series it's better to recommend the whole program and show the pilot. 

Furthermore, there are documentaries and movies. But these sometimes have embedded subtitles, which makes it more difficult.

Also get a distinction between national and foreign movies (western, etc.).

```python
#https://www.npostart.nl/pauw/17-12-2018/BV_101389513
mid = "BV_101389513"
r = requests.get(poms_endpoint + mid + "/nl_NL/CAPTION.vtt")
subtext = r.text
subtext.split(' --> ')
```
